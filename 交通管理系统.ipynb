```python
# 找到一条从start到end的路径
def findPath(graph,start,end,path=[]):   
    path = path + [start]
    if start == end:
        return path 
    for node in graph[start]:
        if node not in path:
            newpath = findPath(graph,node,end,path)
            if newpath:
                return newpath
    return None
```

```python
# 找到所有从start到end的路径
def findAllPath(graph,start,end,path=[]):
    path = path +[start]
    if start == end:
        return [path]
 
    paths = [] #存储所有路径    
    for node in graph[start]:
        if node not in path:
            newpaths = findAllPath(graph,node,end,path) 
            for newpath in newpaths:
                paths.append(newpath)
    return paths
```

```python
# 查找最短路径
def findShortestPath(graph,start,end,path=[]):
    path = path +[start]
    if start == end:
        return path
    shortestPath = []
    for node in graph[start]:
        if node not in path:
            newpath = findShortestPath(graph,node,end,path)
            if newpath:
                if not shortestPath or len(newpath)<len(shortestPath):
                    shortestPath = newpath
    return shortestPath
```

```python
'''
根据不同的需求。
1.根据最小平均车流量推荐
2.根据最小路径推荐
3.根据最小车流量推荐

输入：所有路径
输入：最小平均车流量、最小路径、最小车流量
'''
def findPathLength(ls,weight):
    sum_all =[]
    sum_lengths=[]
    sum_flows=[]
    for lists in ls:
        sum_length =0
        sum_flow = 0.0
        for pint in lists:
            current_pint_info = startpath_info.loc[startpath_info['pint']==pint]
            sum_length += current_pint_info['length'].values[0]
            sum_flow +=(current_pint_info['flow'].values[0])*0.8/16
        sum_flows.append(sum_length)
        sum_lengths.append(sum_length)
        sum_all.append(sum_flow/sum_length)
    ls_fast_index =sum_all.index(min(sum_all))
    ls_best_index =sum_lengths.index(min(sum_lengths))
    ls_empty_index =sum_flows.index(min(sum_flows))
    print(ls_fast_index,ls_best_index,ls_empty_index)
    return ls[ls_fast_index],ls[ls_best_index],ls[ls_empty_index]
```

```python
def get_NowDate(nowdate):
    today = dt.datetime.strptime(nowdate,"%Y-%m-%d")
    weekday = today.weekday() + 1
    if weekday % 7 ==0:
        week = '日'
    elif weekday % 7 ==1:
        week = '一'
    elif weekday % 7 ==2:
        week = '二'
    elif weekday % 7 ==3:
        week = '三'
    elif weekday % 7 ==4:
        week = '四'
    elif weekday % 7 ==5:
        week = '五'
    elif weekday % 7 ==6:
        week = '六'
    print('今天{today}{time},星期{weekday}'.format(today=str(today)[:10],time =str(today)[10:], weekday=week))
    return today
```

```python
def get_NowWeek(nowdate):
    today = dt.datetime.strptime(nowdate,"%Y-%m-%d")
    weekday = today.weekday()
    start = today - dt.timedelta(days=weekday)
    print('这周一是{start},到今天{today}'.format(start=str(start)[:10],today=str(today)[:10]))
    return str(start)[:10],str(today)[:10]
```

```python
def get_LastWeek(nowdate):
    today = dt.datetime.strptime(nowdate,"%Y-%m-%d")
    weekday = today.weekday()
    start = today - dt.timedelta(days=weekday+7)
    end = today - dt.timedelta(days=weekday+1)
    print('这上周一是{start},上周末是{end}'.format(start=str(start)[:10], end=str(end)[:10]))
    return str(start)[:10],str(end)[:10]
```
